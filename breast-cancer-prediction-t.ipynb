**2 ) . Data Set Information:**


Features are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They describe characteristics of the cell nuclei present in the image. A few of the images can be found at
Separating plane described above was obtained using Multisurface Method-Tree (MSM-T) [K. P. Bennett, "Decision Tree Construction Via Linear Programming." Proceedings of the 4th Midwest Artificial Intelligence and Cognitive Science Society, pp. 97-101, 1992], a classification method which uses linear programming to construct a decision tree. Relevant features were selected using an exhaustive search in the space of 1-4 features and 1-3 separating planes.
The actual linear program used to obtain the separating plane in the 3-dimensional space is that described in: [K. P. Bennett and O. L. Mangasarian: "Robust Linear Programming Discrimination of Two Linearly Inseparable Sets", Optimization Methods and Software 1, 1992, 23-34].

> **Ten real-valued features are computed for each cell nucleus:**

    ==============================================


a) radius (mean of distances from center to points on the perimeter)

b) texture (standard deviation of gray-scale values)

c) perimeter

d) area

e) smoothness (local variation in radius lengths)

f) compactness (perimeter^2 / area - 1.0)

g) concavity (severity of concave portions of the contour)

h) concave points (number of concave portions of the contour)

i) symmetry

j) fractal dimension ("coastline approximation" - 1)

```python
import numpy as np
import pandas as pd

# Assuming 'df' is your dataframe and you've already mapped 'diagnosis' to 1/0
# Step 1: Separate features (X) and target (y)
y = df['diagnosis']
X = df.drop(columns=['diagnosis'])

# Step 2: Calculate the correlation matrix (taking absolute values)
corr_matrix = X.corr().abs()

# Step 3: Select only the upper triangle of the matrix
# We do this so we don't drop BOTH features in a correlated pair, just one of them.
upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Step 4: Find all columns where the correlation is greater than 0.90
threshold = 0.90
columns_to_drop = [column for column in upper_triangle.columns if any(upper_triangle[column] > threshold)]

print(f"Features automatically identified to drop: {columns_to_drop}")

# Step 5: Drop those columns from your feature set
X_reduced = X.drop(columns=columns_to_drop)

print(f"\nOriginal number of features: {X.shape[1]}")
print(f"New number of features: {X_reduced.shape[1]}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Step 1: Split the data (80% for training, 20% for testing)
# random_state ensures you get the exact same split every time you run the code
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size=0.2, random_state=42)

# Step 2: Initialize and Train the Random Forest
# n_estimators=100 means the "forest" will be made of 100 individual decision trees
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Step 3: Make predictions on the test set that the model has never seen before
y_pred = rf_model.predict(X_test)

# Step 4: Evaluate the results
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import pandas as pd

# Extract feature importances from the trained model
importances = rf_model.feature_importances_

# Create a DataFrame to view them easily
feature_importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': importances
})

# Sort the features by how important they are
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Display the top 10 most important features
print(feature_importance_df.head(10))
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier

# ==========================================
# PART 1: THE THRESHOLD ADJUSTMENT
# ==========================================

# 1. Get raw probabilities for the 'M' class (index 1)
y_probs = rf_model.predict_proba(X_test)[:, 1]

# 2. Apply the 0.30 threshold
# We use np.where to explicitly map True to 'M' and False to 'B'
custom_threshold = 0.30
y_pred_high_recall = np.where(y_probs >= custom_threshold, 'M', 'B')

# 3. Evaluate the threshold adjustments
print("--- NEW THRESHOLD (0.30) RESULTS ---")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_high_recall))
print("\nClassification Report:\n", classification_report(y_test, y_pred_high_recall))


# ==========================================
# PART 2: THE WEIGHTED MODEL
# ==========================================

# 1. Train the weighted model using the correct string labels ('B' and 'M')
rf_weighted = RandomForestClassifier(n_estimators=100, class_weight={'B': 1, 'M': 3}, random_state=42)
rf_weighted.fit(X_train, y_train)

# 2. Make predictions 
y_pred_weighted = rf_weighted.predict(X_test)

# 3. Evaluate the weighted model
print("\n\n--- WEIGHTED MODEL RESULTS ---")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_weighted))
print("\nClassification Report:\n", classification_report(y_test, y_pred_weighted))
```

```python
%%writefile app.py
import streamlit as st
import pickle
import numpy as np
import pandas as pd

# Set up the local page
st.title("🩺 Breast Cancer Detection App")
st.write("Local deployment: Enter cell measurements below.")

# Load the saved model
@st.cache_resource
def load_model():
    with open('breast_cancer_rf_model.pkl', 'rb') as file:
        return pickle.load(file)

model = load_model()

# Create the user interface for the 20 features
st.sidebar.header("Input Cell Measurements")
feature_names = [
    'radius_mean', 'texture_mean', 'smoothness_mean', 'compactness_mean', 
    'concavity_mean', 'symmetry_mean', 'fractal_dimension_mean', 
    'radius_se', 'texture_se', 'smoothness_se', 'compactness_se', 
    'concavity_se', 'symmetry_se', 'fractal_dimension_se', 
    'smoothness_worst', 'compactness_worst', 'concavity_worst', 
    'symmetry_worst', 'fractal_dimension_worst', 'fractal_dimension_se' 
]

user_input = {}
for feature in feature_names:
    user_input[feature] = st.sidebar.number_input(f"{feature}", value=0.00, format="%.4f")

# Make the Prediction
if st.button("Predict Diagnosis"):
    input_df = pd.DataFrame([user_input])
    probability = model.predict_proba(input_df)[0][1]
    threshold = 0.30 # Our optimized threshold for 100% recall
    
    st.markdown("---")
    if probability >= threshold:
        st.error(f"**Diagnosis: Malignant** 🚨")
    else:
        st.success(f"**Diagnosis: Benign** ✅")
```
